Under er de m√•lte resultatene fra frekvensene vi testet

```python

FREKVENS = np.array([10, 100, 1000, 5000, 10000, 50000, 100000, 500000]) # [Hz]
AMP_CHAN1 = np.array([5.016, 5.027, 4.949, 4.341, 3.348, 0.931, 0.475, 0.099]) # [V]
AMP_CHAN2 = np.array([5.0078, 4.998, 4.974, 4.942, 4.890, 4.831, 4.836, 4.846]) # [V]
FASE_FORSYVNING = np.array([0, 0.2, 6, 28, 46, 82, 91, 92]) # [grader]

forsterkning = AMP_CHAN1 / AMP_CHAN2
print(forsterkning)

forsterkning_db = 20 * np.log10(forsterkning)
print(forsterkning_db)

fig, (ax1, ax2) = plt.subplots(2)
fig.set_size_inches(7, 10)

ax1.set_xscale("log")
ax1.set_xlabel("Frekvens [Hz]")
ax1.set_ylabel("Forsterkning [dB]")
ax1.plot(FREKVENS, forsterkning_db)

ax2.set_xscale("log")
ax2.set_xlabel("Frekvens [Hz]")
ax2.set_ylabel("Fase forsyvning [¬∞]")
ax2.plot(FREKVENS, FASE_FORSYVNING)
```

## Oppgave f

```python
file_names = ["10kHz_CHAN1.csv", "10kHz_CHAN2_REF.csv"]

fig, axs = plt.subplots(len(file_names))

fig.set_size_inches(7, 7)

for i in range(len(axs)):
    file_name = file_names[i]
    ax = axs[i]
    
    data = pd.read_csv(file_name)

    time = data.iloc[:,0].to_numpy()
    signal = data.iloc[:,1].to_numpy()

    ax.plot(time, signal)
    ax.set_title(file_name)

```

```python
fig, ax = plt.subplots(1)

CHAN1_data = pd.read_csv(file_names[0])
CHAN2_data = pd.read_csv(file_names[1])

CHAN1_time   = CHAN1_data.iloc[:, 0].to_numpy()
CHAN1_signal = CHAN1_data.iloc[:, 1].to_numpy()

CHAN2_time   = CHAN2_data.iloc[:, 0].to_numpy()
CHAN2_signal = CHAN2_data.iloc[:, 1].to_numpy()

ax.plot(CHAN1_time, CHAN1_signal, label="CHAN1")
ax.plot(CHAN2_time, CHAN2_signal, label="CHAN2")

ax.legend()
ax.set_xlabel("Tid [s]")
ax.set_ylabel("Spenning [V]")
```
